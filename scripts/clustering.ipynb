```python
import pandas as pd

csv_file = "../data/data_selected_v1.csv"

df = pd.read_csv(csv_file)
df.columns
df.describe()
```

```python
pop_df = df[df['Genre'] == 'pop']
pop_features = pop_df[['Danceability', 'Energy', 'Key', 'Loudness',
       'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness',
       'Valence', 'Tempo', 'Duration_ms', 'Views', 'Likes', 'Comments',
       'Stream']]

#pop_features = pop_df[['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness']]


correlation_matrix = pop_features.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Korrelationsmatrix für das Genre "Pop"')
plt.show()
```

```python
pairplot = sns.pairplot(pop_features, plot_kws={'s': 5})
pairplot.savefig('pairplot_pop.png')
```

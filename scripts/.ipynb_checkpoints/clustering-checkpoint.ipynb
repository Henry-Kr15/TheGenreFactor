```python
import seaborn as sns
import matplotlib.pyplot as plt

csv_file = "../data/data.csv"
df = pd.read_csv(csv_file)
Kpop_df = df[df['Genre'] == 'K-pop']
Jpop_df = df[df['Genre'] == 'J-pop']
synthpop_df = df[df['Genre'] == 'synth-pop']

features = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness']

Kpop_features = Kpop_df[features]
Jpop_features = Jpop_df[features]
synthpop_features = synthpop_df[features]

# DataFrame für den Pairplot erstellen
pairplot_data = pd.concat([Kpop_features, Jpop_features, synthpop_features])
genre_labels = ['K-pop'] * len(Kpop_features) + ['J-pop'] * len(Jpop_features) + ['synth-pop'] * len(synthpop_features)
pairplot_data['Genre'] = genre_labels

# Pairplot erstellen
pairplot = sns.pairplot(pairplot_data, hue='Genre', plot_kws={'s': 5})

# Plot anzeigen
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

csv_file = "../data/data_selected.csv"
df = pd.read_csv(csv_file)

genres = df['Genre'].unique()

genre_data = []
for genre in genres:
    genre_df = df[df['Genre'] == genre]
    genre_features = genre_df[features]
    genre_data.append(genre_features)

# DataFrame für den Pairplot erstellen
pairplot_data = pd.concat(genre_data)
genre_labels = []
for i, genre_features in enumerate(genre_data):
    genre_labels.extend([genres[i]] * len(genre_features))
pairplot_data['Genre'] = genre_labels

# Pairplot erstellen
pairplot = sns.pairplot(pairplot_data, hue='Genre', plot_kws={'s': 5})

# Plot anzeigen
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

csv_file = "../data/data.csv"
df = pd.read_csv(csv_file)

genres = df['Genre'].unique()

selected_genres = ['pop', 'rock', 'metal', 'punk']  # Liste der gewünschten Genres

for genre in selected_genres:
    genre_data = []
    for g in genres:
        if genre.lower() in g.lower():
            genre_df = df[df['Genre'] == g]
            genre_features = genre_df[features]
            genre_data.append(genre_features)

    # DataFrame für den Pairplot erstellen
    pairplot_data = pd.concat(genre_data)
    genre_labels = []
    for i, genre_features in enumerate(genre_data):
        genre_labels.extend([genres[i]] * len(genre_features))
    pairplot_data['Genre'] = genre_labels

    # Pairplot erstellen
    pairplot = sns.pairplot(pairplot_data, hue='Genre', plot_kws={'s': 5})
    plt.show()
```

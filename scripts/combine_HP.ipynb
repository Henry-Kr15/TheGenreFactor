# Werte aus verschiedenen HPOs werden kombiniert

```python
import pandas as pd

df1= pd.read_csv('../HPO-data/optimization_results_chris_2707.csv')
activation_mapping = {'relu': 0, 'LeakyReLU': 1}
df1['activation_numerical'] = df1['activation_function'].map(activation_mapping)
optimizer_mapping = {'adam': 0, 'nadam': 1}
df1['optimizer_numerical'] = df1['optimizer'].map(optimizer_mapping)
df1['epochs'] = 60
df1['weight_contraint'] = 10

df1 = df1.drop(['best_val_loss','val_loss_std','best_train_loss','train_loss_std'], axis=1)

df1 = df1.rename(columns={'activation_function': 'activation',
                          'best_val_accuracy': 'mean_val_score',
                          'best_train_accuracy': 'mean_train_score',
                          'val_accuracy_std': 'std_val_score',
                          'train_accuracy_std': 'std_train_score',
                          'units_1': 'neurons_1', 
                          'units_2': 'neurons_2',
                          'units_3': 'neurons_3',
                          'units_4': 'neurons_4',
                          'dropout_rate': 'dropout'
                         })

df1 = df1.sort_index(axis=1)

print(df1.columns.tolist())
df1.head()
```

```python
df2= pd.read_csv('../HPO-data/optimization_results_final.csv')
activation_mapping = {'relu': 0, 'LeakyReLU': 1}
df2['activation_numerical'] = df2['activation_function'].map(activation_mapping)
df2['optimizer'] = 'adam'
optimizer_mapping = {'adam': 0, 'nadam': 1}
df2['optimizer_numerical'] = df2['optimizer'].map(optimizer_mapping)
df2['epochs'] = 60
df2['weight_contraint'] = 10

df2 = df2.drop(['best_val_loss','val_loss_std','best_train_loss','train_loss_std'], axis=1)

df2 = df2.rename(columns={'activation_function': 'activation',
                          'best_val_accuracy': 'mean_val_score',
                          'best_train_accuracy': 'mean_train_score',
                          'val_accuracy_std': 'std_val_score',
                          'train_accuracy_std': 'std_train_score',
                          'units_1': 'neurons_1', 
                          'units_2': 'neurons_2',
                          'units_3': 'neurons_3',
                          'units_4': 'neurons_4',
                          'dropout_rate': 'dropout'
                         })

df2 = df2.sort_index(axis=1)

print(df2.columns.tolist())
df2.head()
```

```python
df3= pd.read_csv('../HPO-data/optimization_results_v2_test_by_christopher_score.csv')

df3['reduce_lr_factor'] = 0
df3['reduce_lr_min_lr'] = 0
df3['reduce_lr_patience'] = 0
df3['early_stopping_patience'] =0


df3 = df3.drop(['Christopher-Score','split0_test_score', 'split1_test_score', 'split2_test_score','split0_train_score', 'split1_train_score', 'split2_train_score', 'params','mean_fit_time','std_fit_time','mean_score_time','std_score_time', 'rank_test_score'], axis=1)

df3 = df3.rename(columns={'param_batch_size': 'batch_size',
                          'param_epochs': 'epochs',
                          'param_model__activation': 'activation',
                          'param_model__dropout_rate': 'dropout',
                          'param_model__neurons_1': 'neurons_1', 
                          'param_model__neurons_2': 'neurons_2',
                          'param_model__neurons_3': 'neurons_3',
                          'param_model__neurons_4': 'neurons_4',
                          'param_model__num_hidden_layers': 'num_hidden_layers',
                          'param_model__optimizer': 'optimizer',
                          'param_model__weight_constraint': 'weight_contraint',
                          'param_model__activation_numerical': 'activation_numerical',
                          'param_model__optimizer_numerical': 'optimizer_numerical',
                          'mean_test_score': 'mean_val_score',
                          'std_test_score': 'std_val_score'
                         })

df3 = df3.sort_index(axis=1)

print(df3.columns.tolist())
df3.head()
```

```python
df4= pd.read_csv('../HPO-data/optimization_results_v2_test_by_val_acc.csv')

df4['reduce_lr_factor'] = 0
df4['reduce_lr_min_lr'] = 0
df4['reduce_lr_patience'] = 0
df4['early_stopping_patience'] =0

df4 = df4.drop(['split0_test_score', 'split1_test_score', 'split2_test_score','split0_train_score', 'split1_train_score', 'split2_train_score', 'params','mean_fit_time','std_fit_time','mean_score_time','std_score_time', 'rank_test_score'], axis=1)

df4 = df4.rename(columns={'param_batch_size': 'batch_size',
                          'param_epochs': 'epochs',
                          'param_model__activation': 'activation',
                          'param_model__dropout_rate': 'dropout',
                          'param_model__neurons_1': 'neurons_1', 
                          'param_model__neurons_2': 'neurons_2',
                          'param_model__neurons_3': 'neurons_3',
                          'param_model__neurons_4': 'neurons_4',
                          'param_model__num_hidden_layers': 'num_hidden_layers',
                          'param_model__optimizer': 'optimizer',
                          'param_model__weight_constraint': 'weight_contraint',
                          'param_model__activation_numerical': 'activation_numerical',
                          'param_model__optimizer_numerical': 'optimizer_numerical',
                          'mean_test_score': 'mean_val_score',
                          'std_test_score': 'std_val_score'
                         })

activation_mapping = {'relu': 0}
df4['activation_numerical'] = df4['activation'].map(activation_mapping)
optimizer_mapping = {'adam': 0, 'nadam': 1}
df4['optimizer_numerical'] = df4['optimizer'].map(optimizer_mapping)

df4['delta_acc'] = (df4['mean_train_score'] - df4['mean_val_score'])/df4['mean_val_score']

df4 = df4.sort_index(axis=1)

print(df4.columns.tolist())

df4.head()
```

```python
unique_activation = df4['activation'].unique()
print(unique_activation)

unique_optimizer = df4['optimizer'].unique()
print(unique_optimizer)
```

```python
if df1.columns.equals(df2.columns) and df1.columns.equals(df3.columns) and df1.columns.equals(df4.columns):
    combined_df = pd.concat([df1, df2, df3, df4])
else:
    print("Die DataFrames haben nicht die gleichen Spalten.")
```

```python
# Liste der zu überprüfenden Spalten
cols_to_check = ['activation', 'activation_numerical', 'batch_size', 'dropout', 'early_stopping_patience', 'epochs', 'neurons_1', 'neurons_2', 'neurons_3', 'neurons_4', 'num_hidden_layers', 'optimizer', 'optimizer_numerical', 'reduce_lr_factor', 'reduce_lr_min_lr', 'reduce_lr_patience', 'weight_contraint']

# Iteriere über die zu überprüfenden Spalten
for col in cols_to_check:
    # Wenn alle Werte in der Spalte gleich sind
    if combined_df[col].nunique() == 1:
        # Lösche die Spalte
        combined_df = combined_df.drop(col, axis=1)
```
